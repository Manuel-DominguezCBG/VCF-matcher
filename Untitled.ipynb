```python
import pandas as pd
import os
import subprocess
import numpy as np
```

```python
# 1. Remove the header, filter and select data needed
## Remove the header
## File A
os.system("sed '/^##/ d'  W2008872_S8-J4PJL.vcf >  FileA.txt")

## File B 
os.system("sed '/^##/ d'  W2103014_S10-JGFJG.vcf >  FileB.txt")
```

```python
## Convert the files in a pndas data frame
dataA = pd.read_csv("/Users/monkiky/Desktop/Do-these-samples-belong-to-the-same-patient/FileA.txt", delimiter = "\t" )
dataB = pd.read_csv("/Users/monkiky/Desktop/Do-these-samples-belong-to-the-same-patient/FileB.txt", delimiter = "\t" )
dataA
```

```python
## Filter the variants  FILTER = PASS and Variant Frecuency =>0.4

# VF is in the 9th column, we need to extract and put VF data in an independent column first
df2 = dataA.iloc[:,9].str.split(':', expand=True)
df2.columns = dataA.FORMAT.iloc[0].split(':')
dataA = pd.concat([dataA, df2], axis=1)


# Same for dataB
df2 = dataB.iloc[:,9].str.split(':', expand=True)
df2.columns = dataB.FORMAT.iloc[0].split(':')
dataB = pd.concat([dataB, df2], axis=1)

# Results

dataA.head(1)
```

```python
# object to float
dataA["VF"] = dataA.VF.astype(float)
dataB["VF"] = dataB.VF.astype(float)
```

```python
# ok, now we can filter and take the variant we are interested 

dataA = dataA.query('FILTER == "PASS" and VF > 0.4')
dataB = dataB.query('FILTER == "PASS" and VF > 0.4')

```

```python
dataB.head(1)
```

```python
# Let's create Ben's variant identification

#CHROM+POS+RED+ALT
dataA['CHROMPOSREDALT']=dataA["#CHROM"]+dataA["POS"].apply(str)+dataA["REF"]+dataA["ALT"]
dataB['CHROMPOSREDALT']=dataB["#CHROM"]+dataB["POS"].apply(str)+dataB["REF"]+dataB["ALT"]

# Results
dataA
```

```python
dataB
```

```python
df3 = dataA[['CHROMPOSREDALT', 'GT']].copy()
df4 = dataB[['CHROMPOSREDALT', 'GT']].copy()
frames = [df3,df4]
df5 = pd.concat(frames)
df5
```

```python
df5=(df5.assign(key=df5.groupby('CHROMPOSREDALT').cumcount())
      .pivot('CHROMPOSREDALT','key','GT')
      .rename(columns=lambda x:f"Sample{x+1}")
      .rename_axis(columns=None).reset_index())
```

```python
df5["equal?"] = np.where(df5["Sample1"] == df5["Sample2"], True, False)
df5
```
